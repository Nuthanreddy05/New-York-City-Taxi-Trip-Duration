### Employ geographic visualizations to represent pickup and drop-off location 

```python
import folium
pickup_dropoff_map = folium.Map(location=[df['pickup_latitude'].mean(), df['pickup_longitude'].mean()], zoom_start=10)
pickup_locations = folium.FeatureGroup(name='Pickup Locations')
dropoff_locations = folium.FeatureGroup(name='Drop-off Locations')
for index, row in df.iterrows():
    pickup_location = [row['pickup_latitude'], row['pickup_longitude']]
    dropoff_location = [row['dropoff_latitude'], row['dropoff_longitude']]
    folium.CircleMarker(
        location=pickup_location,
        radius=5,
        color='blue',
        fill=True,
    ).add_to(pickup_locations)
    folium.CircleMarker(
        location=dropoff_location,
        radius=5,
        color='red',
        fill=True,
    ).add_to(dropoff_locations)
    folium.PolyLine([pickup_location, dropoff_location], color='green').add_to(pickup_dropoff_map)
pickup_locations.add_to(pickup_dropoff_map)
dropoff_locations.add_to(pickup_dropoff_map)
folium.LayerControl().add_to(pickup_dropoff_map)
pickup_dropoff_map
```

# 4.Algorithm and Data Structure Efficiency

Linkedlist data structure is the best data structure for insertions of new elements at the end of sorted list without the need to shift existing element and more efficient in sorting data.
The nodes in a linked list contain references to the elements that comesÂ after it, it is easy to maintain in  the sorted order.
When new trips are added frequently, appending to the end of a linked list has a constant time complexity, O(1).
linkedlist are dynamic in size where there is no limit on the number elements are inserted in the list.
Linked lists can be easily sorted by adjusting the links between nodes without the need for data movement.
where the size of the dataset is not known before the time, linked lists may be more memory-efficient than arrays.
Each new node in a linked list is placed separately, and the entire data does not need to be shifted.


```python
#example of LinkedList 
class Node:
    def __init__(self,data,trip):
        self.data = data
        self.trip = trip
        self.next = None

class LinkedList:
    def __init__(self):
        self.head = None

    def insert(self,data,trip):
        new_node =Node(data,trip)
        if not self.head or data < self.head.data:
            new_node.next = self.head
            self.head = new_node
        else:
            n = self.head
            while n.next and n.next.data <= data:
                n = n.next
            new_node.next = n.next
            n.next = new_node

    def display(self):
        n = self.head
        while n:
            print(f"trip duration in mins: {n.data} ,trip: {n.trip}")
            n = n.next

sorted_trips = LinkedList()
sorted_trips.insert(30,"1")
sorted_trips.insert(26,"2")
sorted_trips.insert(43,"3")
sorted_trips.display()
```

When new data (a phone number) is added and a goal is to fast filter trips by specific passenger, hash tables perform well.

Hash tables are good for obtaining data based on a given key because they offer constant-time average-case complexity for search operations without looking the entire data.

Hash table provides uniqueness in keys as there no same phone numbers in the data and provides the data of the phone number.

The Hash Table allows for direct access to the data associated with a specific phone number

when data is updated frequently or new trips are added for various passengers, hash tables work well.

```python
class Hash:
    def __init__(self,data,trip, phone_number):
        self.data = data
        self.trip = trip
        self.phone_number = phone_number

class PhoneNumber:
    def __init__(self):
        self.trips = {}

    def insert(self, data, trip, phone_number):
        new_node = Hash(data, trip, phone_number)
        if phone_number not in self.trips:
            self.trips[phone_number] = []
        self.trips[phone_number].append(new_node)

    def phone_number(self, phone_number):
        return self.trips.get(phone_number, [])
    

p= PhoneNumber()
p.insert(30, "1", "9390691699")
p.insert(26, "2", "9056740911")
p.insert(43, "3", "6720975633")

data_of_phone_number = p.phone_number("9390691699")
for tri in data_of_phone_number:
    print(f"trip duration in mins: {tri.data}, trip: {tri.trip}")
```
